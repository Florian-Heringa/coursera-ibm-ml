# Netflix Data Analysis

In this notebook an analysis is presented that shows how movie rating (PG-13, R) genre, runtime and budget are related to popuarity as scored by IMDb rating.

The dataset used is: [danielgrijalvas - Movie Industry (Kaggle)](https://www.kaggle.com/datasets/danielgrijalvas/movies)

## Data Summary

The data consists of 7668 movies, with 15 columns. The target column is `score`, which is expressed as a float between 0 and 10. There are a total of 4 columns with float values, a single column with an integer (year) and 9 categorical columns. Most of the categorical columns are not useful for this analysis (like country, writer, starr, etc.) so most of them will not be used.

```python
data = pd.read_csv("./data/movies.csv")
data.info()
```

```python
f"There are {data['genre'].nunique()} genres and {data['rating'].nunique()} ratings."
```

## Analysis Objective

This analysis will show how IMDb score is related to `['rating', 'genre', 'runtime', 'budget']`.

First I will setup the data for analysis, performing EDA, feature extraction and feature preparation.

```python
feature_cols = ['rating', 'genre', 'runtime', 'budget', 'score']
data = data[feature_cols]
data = pd.get_dummies(data, drop_first=True)
data = data.dropna()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop('score', axis=1)
y = data['score']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=42)
```

## Model Comparison

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

plt.scatter(y_test, y_pred)
```
